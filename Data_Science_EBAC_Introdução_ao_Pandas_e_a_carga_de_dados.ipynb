# Estruturação de dados - Pandas:

O pandas é uma biblioteca de código aberto em Python amplamente utilizada para análise e manipulação de dados. Ela fornece estruturas de dados de alto desempenho e ferramentas de análise de dados que são essenciais para cientistas de dados, analistas de dados e engenheiros de dados. Além disso, o pandas é uma biblioteca fundamental em muitos fluxos de trabalho de ciência de dados e é frequentemente usado em conjunto com outras bibliotecas populares, como NumPy, Matplotlib e Scikit-Learn.

O Pandas apresenta alguns conceitos chaves, tais como:

1. **Estruturas de Dados Principais:**
    - **DataFrame:** O DataFrame é a estrutura de dados mais importante no pandas. Ele é semelhante a uma tabela de banco de dados ou uma planilha do Excel e é usado para representar dados tabulares.
    - **Série:** A Série é uma estrutura unidimensional semelhante a um array ou uma coluna em um DataFrame.


2. **Funcionalidades de Leitura e Escrita:**
    - O pandas suporta a leitura de dados de uma ampla variedade de fontes, incluindo arquivos CSV, Excel, SQL, JSON, HDF5, e muitos outros formatos.
    - Ele também permite escrever dados em diversos formatos.


3. **Manipulação de Dados:**
    - O pandas oferece uma ampla gama de funções para filtrar, selecionar, agrupar, unir e transformar dados.
    - Ele permite a remoção de valores ausentes (NaN) e a aplicação de funções a colunas ou linhas inteiras.

    
4. **Indexação e Seleção:**
    - Os DataFrames podem ser indexados e selecionados com facilidade usando rótulos de coluna e índices de linha.
    - Suporta indexação booleana para filtragem de dados.


5. **Operações Estatísticas:**
    - O pandas oferece funções estatísticas, como média, mediana, desvio padrão, correlação e muitas outras.
    - Você pode realizar agregações de dados com facilidade.


6. **Visualização de Dados:**
    - Embora o pandas não seja uma biblioteca de visualização de dados, você pode integrá-lo facilmente com bibliotecas de plotagem, como Matplotlib e Seaborn, para criar gráficos e visualizações.


7. **Integração com NumPy:** O pandas é construído em cima do NumPy e, portanto, é compatível com as operações NumPy. Você pode facilmente converter entre DataFrames e arrays NumPy.


8. **Lidando com Dados Temporais:**
    - O pandas possui suporte nativo para lidar com séries temporais e dados de data/hora, tornando-o útil para análise de séries temporais.


9. **Desempenho e Eficiência:**
    - O pandas é otimizado para desempenho e é capaz de lidar com grandes conjuntos de dados de maneira eficiente.

O pandas é uma ferramenta extremamente poderosa e versátil para análise de dados em Python. Se você está envolvido em análise de dados ou ciência de dados, é altamente recomendável aprender a usá-lo, pois ele pode facilitar muito o trabalho com dados tabulares e simplificar tarefas de limpeza, transformação e análise de dados.

# Séries

As séries é um tipo de estrutura array unidemensional, sendo capaz de contar qualquer tipo de dado imerso no Python. Os rótulos que justamente como um rótulo ou selo simbolizam um determinado dado são chamdo de índices (index). Vejamos como podemos criar uma série no Python...

Em sua construção, caracterizam-se da seguinte forma:



```
pd.Serie(data = dados, index = indice)
```

Os dados se constituem como elementos de informação, que serão passados para o pandas, o qual irá gerar um quadro, aqui como está em série, em forma de array desses. O índice, caso não seja passado explicitamente, o Python definirá os índices em numerais, começando no zero e seguindo adiante.  


Os dados que podem estar presentes na série são:

* Listas de Python;
* Dicionários de Python;
* ND-Array;
* Valor escalar.

````python
dados = [20.900, 10.450, 4.180, 20.090, 0]

serie_sem_indice = pd.Series(data = dados)
serie_sem_indice
````

````python
serie_sem_indice[0:4]
````

````python
dados = [20.900, 10.450, 4.180, 20.090, 0]
index = ["A", "B", "C", "D", "E"]

serie = pd.Series(data = dados, index = index)
serie
````

````python
serie["A"]

print("")

serie["A":"D"]
````

````python
popBrasilDict = {"Amazonas": 398336,
                 "São Paulo": 44396484,
                 "Bahia": 15203934,
                 "Minas Gerais": 20869101,
                 "Roraima":505665,
                 "DF": 2914830}

popBrasilSerie = pd.Series(popBrasilDict)
popBrasilSerie
````

````python
dados_estados = pd.DataFrame({"populacao": popBrasilSerie,
              "area": areaBrasilSerie})

dados_estados
````

````python
area = dados_estados["area"]
area
````

**Criando uma nova coluna no Python:**

Para criar uma nova coluna no DataFrame do pandas basta eu fazer como jaz abaixo, ou seja, eu digo o nome do banco de dados no qual eu quero criar a nova coluna, conectando-o com a coluna que desejo criar, depois atribuindo a ela seus determinados e respectivos valores.  

````python
dados_estados["densidade"] = round(dados_estados["populacao"] / dados_estados["area"], 2)

dados_estados
````

````python
def densidade_pop(populacao: float, area: float) -> float:

  densidade_pop = sum(populacao) / sum(area)
  densidade_pop = round(densidade_pop, 2)

  print(f"A densidade populacional por habitantes no Brasil é de {densidade_pop} h/m², em que h se refere ao número de habitantes.")
````

````python
densidade_pop(populacao=pop, area=area)
````

````python
# Antes realizarmos operações aritiméticas com o dataframe, iremos criar um.
# Dado que os dataframes são constituidos por meio de matrizes, afinal, os arrays
# iremos notar que as unicas operações possíveis são: soma, subtração, multiplicação e potenciação.

a = np.random.randint(10, size =(3,5)) # Estou criando um conjunto aleatório com 10 valores
                                       # de tamho 3x5, isto é, 3 linhas e 5 colunas.
a


````

````python
z = np.random.randint(10, size =(1,5))
z

df_z = pd.DataFrame(z, columns = ["A", "B", "C", "D", "E"])
df_z
````

**Soma:**

````python
df_a + df_b

# Podemos somar os valores de um df em relação ao outro, bastanto realizar a operação
# acima. Não obstante, ele seleciona o valor de cada célula respectiva, formada pelas
# sua posição em relação às linhas e colunas, e a soma. Por exemplo, tomando como base
# a primeira linha e a primeira coluna dos df's a e b, temos o valor 4 para o df_a
# e o valor 3 para o df_b, somando-os, chegamos ao valor 7, podendo ser encontrado
# no novo df formado pela soma do df_a e do df_b.
````

````python
df_x = df_a + df_z
df_x
# O NaN expressa um conjunto de valores nos quais não há número disponível,
# isso acontece, pois estamos somando duas matrizes que não possuem o mesmo
# número de linhas, apenas de colunas.
````

````python
# Durante um tratamento de dados, não podemos deixar valores NaN, porque
# além de não ser uma boa prática, não nos permite realizar operações quaisquer.
# Uma boa forma de solucionar esse percalço, é substituir os valores NaN para algum outro,
# geralmente booleano, como 0 ou 1. Aqui estarei preechendo com 0, vejamos como:


# Substituindo os valores NaN por zero
# Traduzindo o código...
# Do df_x, preencha valores NaN com zero. Colocando inplace = True, quero dizer que
# desejo que a substituição ocorra.
df_x.fillna(0, inplace=True)

df_x
````

````python
# Configurando as opções de exibição para mostrar números inteiros
pd.options.display.float_format = '{:.0f}'.format

# Mostrando o DataFrame com valores inteiros
print(df_x)
````

# Trabalhando com DataFrame da internet:

````python
# Eu posso ler um arquivo csv presente na internet, não havendo a necessidade obrigatória
# de baixá-lo
url = "https://s3.sa-east-1.amazonaws.com/ckan.saude.gov.br/SIPNI/COVID/uf/uf%3DAM/part-00004-7af288af-6498-4fa5-89d0-45a49ba82375.c000.csv"

# Note que eu estou colocando algo adicional ao o que normalmente estou acostumado a fazer: estou, além de informar ao pandas o que
# quero ler, o tipo de separação dos itens que o determinado arquivo se encontra. Esse não é separado por ",", mas ";", assim, para
# que eu consiga gerar o dataframe da melhor forma.

df_sus = pd.read_csv(url, sep=";")
````

````python
df_sus_copy = df_sus[:20000].copy() # Copiando apenas alguns valores do conjunto inteiro de linhas,
                                    # sendo no caso 10k
````

````python
df_sus = df_sus_copy
df_sus
````

Mas como podemos olhar para esses dados, digo tirar informação deles? Podemos nos perguntar quantos vacinados foram na capital do estado, mas como fazer?

# Porcentagem de vacinados na capital:

````python
# Queremos encontrar apenas os vacinados de Manaus que é a capital de Amazônas. Uma forma de se fazer isso
# é criar uma coluna na qual seja uma bandeira da capital, isto é, para dados que se relacionam à capital,
# seu valor será 1, indicando o True, enquanto que para os que não seu valor será 0, indicando o False.
# Para tanto, vamos:

# criando de início uma nova coluna com o respectivo nome, a qual indica o valor 0
# para todas as capitais.
df_sus["flag_capital"] = 0
````

````python
df_sus
````

````python
# Agora queremos diferenciar, uma vez que todas as localidades estão com o valor de flag_capital
# igual a 0 e queremos fazer com que os dados que correspondam à capital do Estado seja equivalente a 1,
# como proceder?

df_sus.loc[df_sus["paciente_endereco_nmMunicipio"] == "MANAUS"]
````

````python
df_sus.loc[df_sus["paciente_endereco_nmMunicipio"] == "MANAUS", "flag_capital"] = 1

# O que estou
````

````python
df_sus.head()
````

````python
# Selecionando apenas as duas colunas requisitadas.

df_sus[["paciente_endereco_nmMunicipio", "flag_capital"]].head(10)
````

````python
# Eu pretendo saber o percentual de pessoas, certo? Para tanto, é só eu somar
# a quantidade de vacinados na capital e depois colocá-la na razão com a quantidade
# de dados, que aqui coincide com a quantidade de linhas, desse modo:

total_cidades = len(df_sus) # Lendo o comprimento dos dados.
total_cidades
````

````python
vacinados_na_capital = df_sus["flag_capital"].sum() # Uma vez que eu tenho uma coluna que serve
                                                    # como a bandeira da capital, indicando se a
                                                    # vacinação se deu nela ou não pelos valores 0 e 1
                                                    # respectivamente, ao somar saberei que o total
                                                    # indicará, no conjunto de dados, a porção em que
                                                    # as vacinações ocorreram na capital.
vacinados_na_capital
````

````python
# Calculando o percentual de vacinados na capital e depois arredondando.

percentual_vacinados = (vacinados_na_capital / total_cidades)
percentual_vacinados *=100
percentual_vacinados = round(float(percentual_vacinados), 2)

percentual_vacinados
````
