```python
# Preenchendo o gráfico com dados quaisquer:

# Como primeiro parâmetro insere-se a informação, enquanto que
# como o segundo se relaciona a parte artística do gráfico. No caso,
# para tanto, podemos passar a cor da linha do gráfico que queremos
# elaborar, representada pelo k, bem como a forma dessas, que, uma vez
# que colocamos dois traços, será tracejada, como vemos abaixo.

plt.plot(np.random.randn(50).cumsum(), "k--")
```

```python
# Inicialmente criamos um gráfico que continha quatro subplots, certo?
# Mas há uma forma de fazer isso de modo automatizado, sem que precise
# manualmente fazer eixo a eixo? Sim, a função subplot, que veremos a seguir:

fig, axes = plt.subplots(2,3)
```

**Espaçamento das subplotagens:**

```python
fig, axes = plt.subplots(2,2, sharex=True, sharey=True)

# Dentro desses dois ciclos for, eu executo o processo duas vezes, como
# demonstra o alcance - o range - do ciclo, cada qual para as respectivas i e j.
# Se não percebeu, eu estou formando a partir disso um esquema de matrizes, que
# relaciona os valores que servirão para o eixo x e outro para o eixo y.

for i in range(2):

  for j in range(2):

    axes[i, j].hist(np.random.randn(500), bins=50, color="k", alpha=0.9)

# Ajustando os subplots para que ficam um junto do outro por meio do método
# subplots_adjuts:

plt.subplots_adjust(wspace=0, hspace=0)
```

```python
dt = np.random.randn(30).cumsum()

# Criando um gráfico que armazena "duas funções" diferentes.
# Eu posso, no processo artístico da geração dos gráficos
# especificar as cores que serão passadas, bem como estilo da linha
# e afins de modo explicíto, como fiz a seguir:

plt.plot(dt, color="k", linestyle="--", label="Default")
plt.plot(dt, color="b", linestyle="-", drawstyle="steps-post",label="steps-post")
plt.legend(loc="best")

# Perceba que a linha em azul denota um desenho que "segue" a linha tracejada
# foi possível de fazê-la, pois eu utilizei o método drawstyle, sendo esse estilo
# steps-posts.
```

```python
# Também podemos submeter ao seaborn uma operação dentro da lógica
# do matpllot, vejamos:

# O que significa a função plt.rc( )?

# A função plt.rc( ) é frequentemente usada para modificar configurações padrão
# do Matplotlib antes de criar os gráficos. Então, basicamente, por meio dele
# eu altero as configurações que por padrão seriam de uma determinada forma.
# No caso, estou passando o tipo de plotagem que será, bem como seu tamanho
# esperado, acho que é isso.

plt.rc("figure", figsize=(10,8))

# Criando uma plotagem que terá como base a criação de
# quatro gráficos, dois na porção de cima e os demais embaixo.

fig, axes = plt.subplots(2,2)

# Criando os gráficos propriamente ditos. A partir do sns.countplot
# a operação manual que acima eu tive que realizar com o matplot não
# se faz necessária, uma vez que per se ele já realiza o procedimento
# a mim. Da mesma forma, perceba que eu passo um termo no eixo x e
# logo depois a data, que armazena o conjunto de dados utilizado.

# Eu havia ficado com dúvida sobre como bastava eu especificar o
# que gostaria que estivesse no eixo x, que o gráfico esperado fosse criado.
# Imagino que, para tanto, há uma combinação. Quando especifico o que desejo
# que esteja no eixo x e depois passo o df, ele irá criar um gráfico
# considerando o valor ou termo que fora explicitado, de tal modo que
# esses representarão o "domínio" apresentando cada qual a sua "imagem".

sns.countplot(ax = axes[0,0], x="day", data=df_tips)
sns.countplot(ax = axes[0,1], x="gender", data=df_tips)
sns.countplot(ax = axes[1,0], x="smoker", data=df_tips)
sns.countplot(ax = axes[1,1], x="time", data=df_tips)
```

Podemos, ainda, a partir dos gráficos gerados pelo seaborn encontrarmos a média variável dos dados de apreciação. Para tanto, basta que pedimos, ao invés do countplot - que realiza a contagem - o barplot. Passando também como parâmetro uma coluna do df que representará o eixo y, ele irá realizar o cálculo da média.

```python
# Também podemos usar o barplot - gráficos de barras - que traz a média variável
# y por padrão, vejamos:

plt.rc("figure", figsize=(10,8))

# Criando uma plotagem que terá como base a criação de
# quatro gráficos, dois na porção de cima e os demais embaixo.

fig, axes = plt.subplots(2,2)

# Realizando um gráfico de contagem das gorjetas por dia.
sns.countplot(ax = axes[0,0], x="day", data=df_tips)

# Realizando um gráfico que considera a média das porcentagens das gorjetas
# por dia.
sns.barplot(ax = axes[0,1], x="day", y="tips_pct",data=df_tips)

# Realizando um gráfico que considera a média das gorjetas
# por dia.
sns.barplot(ax = axes[1,0], x="day", y="tip",data=df_tips)

# Realizando um gráfico que considera o montante médio das gorjetas por dia.
sns.barplot(ax = axes[1,1], x="day", y="tip", estimator=sum, data=df_tips)
```

Essa barra na vertical preta em cima do gráfico de barra simboliza o intervalo de confiança, em inglês é confidence interval. Como aprendi nas aulas de estatística com o Shine, é um intervalo que compreende os valores máximos e mínimos de variação de um valor, com uma determinada confiançã, que pode ser de 90%, 95% - o de uso consensual -, de 98% e assim por diante.

Para que eu remova esse intervalo de confiança, basta passar como parâmetro: ci=None, vejamos um exemplo:

```python
plt.rc("figure", figsize=(10,8))

# Criando uma plotagem que terá como base a criação de
# quatro gráficos, dois na porção de cima e os demais embaixo.

fig, axes = plt.subplots(2,2)

# Realizando um gráfico de contagem das gorjetas por dia.
sns.countplot(ax = axes[0,0], x="day", data=df_tips)

# Realizando um gráfico que considera a média das porcentagens das gorjetas
# por dia.
sns.barplot(ax = axes[0,1], x="day", y="tips_pct",data=df_tips)

# Realizando um gráfico que considera a média das gorjetas
# por dia.
sns.barplot(ax = axes[1,0], x="day", y="tip",data=df_tips)

# Realizando um gráfico que considera o montante médio das gorjetas por dia.
sns.barplot(ax = axes[1,1], x="day", y="tip", estimator=sum, data=df_tips, ci=None)
```

Consegue reparar que no último gráfico não há mais as barras como haviam antes? Então, eu removi o intervalo de confiança.

**Gráfico de Pareto:**

O gráfico de Pareto funciona criando um eixo y "irmão", dividindo os dados do eixo x, com o eixo y principal. Por meio dele conseguimos identificar qual a frequência do valor de y, com base numa informação do eixo x. Vejamos como ele funciona:

```python
# Antes de criar o gráfico de Pareto de fato, irei criar um dataset.

df_pizzaria = pd.DataFrame({"contagem":[20,43, 174, 5, 2, 1, 1, 0]})

# Criando os índices - index - que estarão no nosso df.
df_pizzaria.index = ["Produto com defeito", "Produto errado", "Tempo de entrega",
                     "Horário de entrega", "Mau atendimento", "Valor errado",
                     "Não serve", "Desistência"]

# Ordenando a contagem a partir da "contagem", sendo essa realizada de modo
# decrescente, do maior ao menor.
df_pizzaria = df_pizzaria.sort_values(by="contagem", ascending=False)

# Criando uma outra coluna que armazena o montante de cada uma das reclamações.
# Pelo menos foi o que eu entendi.
df_pizzaria["somatório"] = df_pizzaria["contagem"].cumsum() / df_pizzaria["contagem"].sum()*100

df_pizzaria
```

```python
# Elaborando o gráfico de Pareto:]

from matplotlib.ticker import PercentFormatter

fig, axis = plt.subplots()

# Eixo y principal: Conta como domínio os valores de x,
# sendo o valor da imagem os valores da contagem, encontradas
# a partir do momento no qual eu escrevo: df_pizzaria["contagem"].

axis.bar(df_pizzaria.index, df_pizzaria["contagem"])

# Criando um eixo y irmão ao principal.
axis2 = axis.twinx()

# Elaborando o que esse eixo y irmão será, estabelecendo como seu domínio
# os valores do eixo x, como sua imagem os valores de df["somatório"], bem como
# definindo a sua cor e marcador - estilo gráfico utilizado.
axis2.plot(df_pizzaria.index, df_pizzaria["somatório"], color="k", marker="o")

# Para deixar mais legível, eu posso transformar os valores do eixo y "irmão" em
# porcentagem a partir do procedimento abaixo.
axis2.yaxis.set_major_formatter(PercentFormatter())

# Definindo o estilo gráfico dos valores que estarão no eixo x e eixo y "irmão".
axis.tick_params(axis="x", rotation=30)
axis2.tick_params(axis="y", colors="k")

# Exibindo o gráfico.
plt.show()
```

# Seaborn e bivariadas discretas:

Nessa parte iremos trabalhar com a construnção de gráficos que assumem ao mesmo tempo duas entradas. Dentre os tipos gráficos, os que iremos ver no primeiro momento são:



*   sns.countplot();
*   sns.catplot();
*   df.crosstab().



```python
df_tips_sem_outiliers.head()
```

```python
# Analisando a distribuição de fumantes e não fumantes por dia:

# O gráfico gerado irá dividir em seu resultado dois cenários,
# mutuamente excludentes - sim e não -, para cada variável de x,
# que no caso serão os dias. A variável que é tomada em consideração
# para realizar essa segmentação é a smoker.

# Gráfico de fumantes e não fumantes por dia.

sns.countplot(x="day", hue="smoker",
              data=df_tips_sem_outiliers)
```

```python
# Se eu quiser saber quais são os dias de preferência para cada grupo,
# fumantes ou não fumantes, faço:

sns.countplot(x="smoker", hue="day",
              data=df_tips_sem_outiliers)
```

```python
# Tabela Cruzada:

# No gráfico acima estou relacionando a quantidade de fumantes em cada dia,
# porém é comum, na verdade, ao invés de mensurarmos a quantidade em termos
# brutos, mensurar com base no percentual. Para tanto, vejamos:

frequency = pd.crosstab(df_tips_sem_outiliers["smoker"], df_tips_sem_outiliers["day"])
frequency
```

```python
# Aqui temos a quantidade de fumantes e não fumantes com base
# nos respectivos dias. Como a gente quer o percentual dos dias
# devemos criar uma coluna que irá somar todos os dias e depois dividir
# o valor dessa coluna dia a dia, para que tenhamos o percentual.
# Para tanto, vamos:

# somando os valores de cada linha do novo dataframe. Eu coloco dentro, como
# parâmetro da função soma "axis=1", porque eu quero somar as linhas, e não as colunas.

frequency["total"] = frequency.sum(axis=1)
frequency
```

```python
# Criando um novo dataframe no qual compreenda
# em termos percentuais a quantidade de fumantes e não fumantes.
# Para a sua elaboração, utilizo a função div - de division -,
# passando para ela o total, ou universo dos dados, primeiramente
# por aquilo que eu quero descobrir o percentual - sendo o numerador.

frequency_pct = frequency.div(frequency["total"], axis = 0)
frequency_pct
```

```python
# Criando o gráfico de barras, agora estando esse com base no percentual.

frequency_pct.plot.bar()
```

```python
# Entendendo o catplot:

# O catplot é um recurso gráfico que com base nas entradas, que são passadas
# como parâmetro, retorna o valor médio dessas, com um intervalo de confiança.

# Para a sua construnção, basta:

sns.catplot(x="day",
            y="tips_pct",
            kind="bar",
            data=df_tips_sem_outiliers)
```

```python
# Criando um gráfico catplot capaz de me mostrar o total de consumo
# realizado em cada dia por fumantes e não fumantes. A variável hue é
# a maneira pela qual eu consigo segmentar essa análise, em excrutíneo,
# um grupo do outro.

sns.catplot(x="day",
            y="total_bill",
            hue="smoker",
            kind="bar",
            data=df_tips_sem_outiliers,
            ci=None) # Coloquei ci igual a None, pois não queria mostrar
                     # o intervalo de confiança.
```

```python
# Fazendo uma rede de gráficos:

sns.catplot(x="day", y="tips_pct",
          row="time",
            col="smoker",
            hue="gender",
            kind="bar",
            data=df_tips_sem_outiliers)
```

```python
# Histograma com covariáveis:

sns.displot(data=df_pg,
            x="body_mass_g",
            hue="sex",
            element="step")

# A partir desse histograma, conseguimos observar que há, pela densidade dos gráficos
# representada pelos dois pontos de máximos, duas termos principais, que se trata de um
# gráfico bivariado, por assim dizer.

# Por meio dele, percebemos que há dois grupos principais, tanto feminino quanto
# masculino, que apresentam dois picos de massa diferente. Porém, esse gráfico
# apresenta uma função de introdução interessante, necessitando, porém, de outros,
# para que podemos fazer uma análise mais aprofundada. Nesse sentido, podemos
# querer utilizar outros gráficos, como o boxplot ou de dispersão, vejamos como:
```

```python
# Gráfico boxplot:

# Criando um gráfico boxplot. Note que ele despreza os dois contigentes
# que possuiam picos de massa diferentes, mas o soma, produzindo um único
# boxplot relativo a cada sexo.

# Ainda que isso possa ser uma primeira abordagem usual, a sua falta de
# especificidade provoca uma perda de informação, fazendo-se necessário
# "minerar" mais os dados, tais como nos exemplos a seguir.

sns.boxplot(x="sex", y="body_mass_g", data=df_pg)


```

```python
sns.boxplot(x="species", y="body_mass_g", hue="sex", data=df_pg)

# Aqui, especificando as espécies, eu consigo realizar uma análise
# mais apurada, conseguindo identificar as densidades de massa para
# cada espécie, bem como relativo ao sexo.

# Podemos observar que esse boxplot ao contrário dos primeiros
# elaborados, é construído a partir de duas entradas, uma para
# o eixo x e outra para o eixo y, além da varíavel hue, que permite
# segmentar em dois grupos um mesmo elemento, que se constitui pela
# união de dois subgrupos, sendo aqui os sexos masculino e feminino.
```

```python
# Gráficos de dispersão (scaterplot):

# Podemos, ainda, querer analisar os dados sob um gráfico de
# dispersão. Para tanto, usamos o scatterplot, vejamos:

# OBS: A variável alpha se refere ao grau de transparência dos
# dados, de tal modo que nos permite identificar, por meio da coloração
# mais forte do que o padrão, a partir de um alpha baixo, a densidade dos
# dados, seu nível de prevalência.

sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",
                alpha=0.5,
                data=df_pg,
                hue="species")
```

```python
# Considerando também a massa dos bichinhos.

sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",
                alpha=0.5,
                data=df_pg,
                hue="species",
                size="body_mass_g")
```

```python
# Aliando o gráfico de dispersão, mas com a capacidade de prover
# outras informações gráficas, como gráficos de densidade, podemos
# utilizar o jointplot, vejamos:

sns.jointplot(x="bill_length_mm",
                y="bill_depth_mm",
                alpha=0.5,
                data=df_pg,
                hue="species")
```

Quando se tem uma grande porção de dados utilizar apenas um gráfico de dispersão pode ser pouco eficiente. Desse modo, podemos querer realizar uma série de gráficos que nos permita, para cada variável, realizar uma análise.

```python
sns.pairplot(df_pg, hue="species")
```

```python
g = sns.PairGrid(df_pg, hue="species")
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)

# A principal diferença desse gráfico em relação ao de cima são duas:
# Conseguimos ver a linha da função de densidade, bem como notamos, como
# se fosse uma topografia, a densidade dos dados.
```
